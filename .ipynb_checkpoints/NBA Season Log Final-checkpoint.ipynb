# Graph record over time relative to .500 for NBA season

```python
%matplotlib inline
import seaborn as sns
import os

import requests

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import pandas as pd

import datetime
```

### Get team information
This "Team Constants" dataframe has basic information about each team that doesn't change from year to year (for the most part) and includes reference information like name, abbreviation, city, conference, and colors. This will be used to iterate and subset (e.g. to plot each conference separately) as well as for coloring and labeling the charts.

```python
# get list of teams with attributes from nba_py
from nba_py import constants

# get team information
teams = pd.DataFrame.from_dict(constants.TEAMS, orient='index') # as a dataframe, indexed by team abbr
# clean up color column
teams['color'] = teams['color'].map(lambda x: x.lstrip('#'))

teams
# teams.index
```

```python
# nba_api
from nba_api.stats.endpoints import teamgamelog

# function to get the team game log, convert to a df, and calculate record relative to .500
def gameLogByTeam(team_id, season, seasonType):
    # define columns to keep in from the imported data
    cols=['Team_ID','GAME_DATE','WL','W','L']

    # get game log data
    log = teamgamelog.TeamGameLog(team_id,season,seasonType)
    log_df = log.team_game_log.get_data_frame()

    # just keep necessary columns
    df = pd.DataFrame(data=log_df,columns=cols)

    # Calculating column for wins above .500
    df['winsAbove500'] = (df["W"]-df["L"])
    df.columns = ['id','GAME_DATE','WL','W','L','winsAbove500']
    
    df.loc[82] = [team_id,'OCT 16, 2017',"No Game",0,0,0]

    # format date
    df['Date']=pd.to_datetime(df['GAME_DATE'])
    df.set_index('Date')    
    
    return df


# function to get the team game log, convert to a df, and calculate record relative to .500
def gameLogByTeamWithStartDate(team_id, season, seasonType, startDate):
    # define columns to keep in from the imported data
    cols=['Team_ID','GAME_DATE','WL','W','L']

    # get game log data
    log = teamgamelog.TeamGameLog(team_id,season,seasonType)
    log_df = log.team_game_log.get_data_frame()

    # just keep necessary columns
    df = pd.DataFrame(data=log_df,columns=cols)

    # Calculating column for wins above .500
    df['winsAbove500'] = (df["W"]-df["L"])
    df.columns = ['id','GAME_DATE','WL','W','L','winsAbove500']
    
    startDate = startDate
    df.loc[-1] = [team_id, startDate,"No Game",0,0,0]

    # format date
    df['Date']=pd.to_datetime(df['GAME_DATE'])
    df.set_index('Date')    
    
    return df

# function to get the team game log, convert to a df, and calculate record relative to .500
def saveGameLogByTeam(team_id, season, seasonType, team_abbr):
    # define columns to keep in from the imported data
    cols=['Team_ID','GAME_DATE','WL','W','L']

    # get game log data
    log = teamgamelog.TeamGameLog(team_id,season,seasonType)
    log_df = log.team_game_log.get_data_frame()

    # just keep necessary columns
    df = pd.DataFrame(data=log_df,columns=cols)

    # Calculating column for wins above .500
    df['winsAbove500'] = (df["W"]-df["L"])
    df.columns = ['id','GAME_DATE','WL','W','L','winsAbove500']
    
#     df.loc[82] = [team_id,'OCT 16, 2017',"No Game",0,0,0]

    # format date
    df['Date']=pd.to_datetime(df['GAME_DATE'])
    df.set_index('Date')    
    
    file_name = "gameLog_" + team_abbr + "_" + season + seasonType
    
    df.to_csv(file_name)
    
    print("Saved file as: " + file_name + ".csv")
```

```python
# Save game log data locally
season='2017-18'
seasonType = 'Regular Season'

# team_abbr = 'IND'
# team_id = '1610612754'
# CHI_id = '1610612741'

### Had to do this team by team manually because the API was limiting requests to 10 within a loop
### Added the "requests" package and it solved the issue

for i in teams.index:
    if teams.loc[i,'conference'] == 'Eastern':
        
        team_abbr = str(teams.loc[i,'abbr'])
        team_id = str(teams.loc[i,'id'])

        saveGameLogByTeam(team_id, season, seasonType, team_abbr)
```

```python
### Plot function

def plotGameLog(gamelog_df, team_abbr, team_color, label_date):
    df = gamelog_df
     
    label_x_pos = label_date
    label_y_pos = df['winsAbove500'][0]
    record_text = str(df['W'][0]) + "-" + str(df['L'][0])
    label_text = team_abbr + " " + record_text
    
#     print(label_x_pos, label_y_pos, label_text)
    
    # plot each team
    plt.step(df['Date'], df['winsAbove500'], color = team_color)
    plt.text(label_x_pos, label_y_pos, label_text, fontsize=10, color=team_color) 
```

```python
# setting universal variables for API call and formatting
season='2017-18'
seasonType = 'Regular Season'

label_date = datetime.datetime(2018, 4, 12, 0, 0)


# loop through each team
for i in teams.index:
    if teams.loc[i,'conference'] == 'Western':
        
#         print(row["id"], row['abbr'], row['color'])
        team_id = str(teams.loc[i,'id']) # set team id for API call
        team_color = "#{}".format((teams.loc[i,'color']))
        team_abbr = str(teams.loc[i,'abbr'])
        
        print(team_id, team_color, team_abbr)
        
        plotGameLog(gameLogByTeam(team_id, season, seasonType),team_abbr, team_color, label_date)
```

```python
### Set up plot area - Eastern Conference

plt.figure(figsize=(15, 12))



ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)  

ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()  

ax.grid(linestyle='-', alpha = 0.3)

# setting universal variables for API call and formatting
season='2017-18'
seasonType = 'Regular Season'
label_date = datetime.datetime(2018, 4, 12, 0, 0)

# loop through each team
for i in teams.index:
    if teams.loc[i,'conference'] == 'Eastern':
        
#         print(row["id"], row['abbr'], row['color'])
        team_id = str(teams.loc[i,'id']) # set team id for API call
        team_color = "#{}".format((teams.loc[i,'color']))
        team_abbr = str(teams.loc[i,'abbr'])
        
#         print(team_id, team_color, team_abbr)
        
        plotGameLog(gameLogByTeam(team_id, season, seasonType),team_abbr, team_color, label_date)
        

plt.savefig("EasternConference_GamesAbove500_201718.svg", format="svg")
```
