```python
# testing nba_api
from nba_api.stats.endpoints import teamgamelog

def gamelogByTeam(team_id, season, seasonType):
    # define columns to keep in from the imported data
    cols=['Team_ID','GAME_DATE','WL','W','L']

    # get game log data
    log = teamgamelog.TeamGameLog(team_id,season,seasonType)
    log_df = log.team_game_log.get_data_frame()

    # just keep necessary columns
    df = pd.DataFrame(data=log_df,columns=cols)

    # Calculating column for wins above .500
    df['winsAbove500'] = (df["W"]-df["L"])
    df.columns = ['id','GAME_DATE','WL','W','L','winsAbove500']
    
    df.loc[82] = [team_id,'OCT 16, 2017',"No Game",0,0,0]

    # format date
    df['Date']=pd.to_datetime(df['GAME_DATE'])
    df.set_index('Date')    
    
    return df
    
```
